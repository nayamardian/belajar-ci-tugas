```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
```

```python
# baca dataset
df = pd.read_csv('/content/drive/MyDrive/datasets/jmlpenduduk-2022_with_missing_value(1).csv')

# tampilkan 5 baris pertama
df.head()
```

```python
# liat struktur data
df.info()

# cek jumlah missing value per kolom
print("\nMissing value per kolom:")
print(df.isnull().sum())

# biar lebih jelas, tampilkan baris yg ada missing value
df[df.isnull().any(axis=1)]
```

```python
# isi missing value secara logis
df['Total'] = df['Total'].fillna(df['Laki-laki'] + df['Perempuan'])
df['Laki-laki'] = df['Laki-laki'].fillna(df['Total'] - df['Perempuan'])
df['Perempuan'] = df['Perempuan'].fillna(df['Total'] - df['Laki-laki'])
```

```python
print("Missing value setelah diisi:")
print(df.isnull().sum())

df.head(10)
```

```python
df.to_csv('/content/drive/MyDrive/datasets/penduduk_clean.csv', index=False)
print("dataset bersih udah disimpan ke drive ")
```
